# Load Data

```python
def load_mushroom_dataset() -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Load the mushroom dataset from UCI ML Repository.

    Returns:
        Tuple[pd.DataFrame, pd.DataFrame]: Features (X) and target (y) as pandas DataFrames
    """
    try:
        # Import ucimlrepo library
        from ucimlrepo import fetch_ucirepo

        logger.info("Fetching Mushroom dataset from UCI repository...")
        mushroom = fetch_ucirepo(id=73)

        # Extract features and target
        X = mushroom.data.features
        y = mushroom.data.targets

        logger.info(f"Dataset loaded successfully with {X.shape[0]} samples and {X.shape[1]} features")

        # Display basic dataset information
        print("Dataset metadata:")
        print(f"Name: {mushroom.metadata.name}")
        print(f"Number of instances: {mushroom.metadata.num_instances}")
        print(f"Number of features: {mushroom.metadata.num_features}")
        print(f"Feature types: {mushroom.metadata.feature_types}")

        return X, y

    except Exception as e:
        logger.error(f"Error loading dataset: {str(e)}")
        raise
```

## Handling Missing Value

```python
def handle_missing_values(X: pd.DataFrame, strategy: str = 'special_value') -> pd.DataFrame:
    """
    Handle missing values in the dataset using the specified strategy.

    Args:
        X (pd.DataFrame): Feature dataframe
        strategy (str): Strategy for handling missing values
            - 'special_value': Replace with a special category 'missing'
            - 'drop': Drop rows with missing values
            - 'most_frequent': Replace with most frequent value

    Returns:
        pd.DataFrame: DataFrame with handled missing values
    """
    logger.info(f"Handling missing values using strategy: {strategy}")

    # Create a copy to avoid modifying the original
    X_processed = X.copy()

    # Find columns with missing values
    missing_cols = X_processed.columns[X_processed.isnull().any()].tolist()

    if not missing_cols:
        print("No missing values to handle.")
        return X_processed

    print(f"Columns with missing values: {missing_cols}")

    if strategy == 'drop':
        # Drop rows with missing values
        initial_rows = len(X_processed)
        X_processed = X_processed.dropna()
        final_rows = len(X_processed)
        print(f"Dropped {initial_rows - final_rows} rows with missing values")
        print(f"Remaining samples: {final_rows}")

    elif strategy == 'most_frequent':
        # Fill with most frequent value
        for col in missing_cols:
            most_freq = X_processed[col].mode()[0]
            X_processed[col].fillna(most_freq, inplace=True)
            print(f"Filled missing values in '{col}' with most frequent value: '{most_freq}'")

    elif strategy == 'special_value':
        # Fill with a special value 'missing'
        for col in missing_cols:
            X_processed[col].fillna('missing', inplace=True)
            print(f"Filled missing values in '{col}' with 'missing'")

    # Verify no missing values remain
    remaining_missing = X_processed.isnull().sum().sum()
    print(f"Missing values remaining: {remaining_missing}")

    return X_processed
```

## Label Encoding

```python
def encode_categorical_features(X: pd.DataFrame, y: pd.DataFrame = None) -> Tuple[np.ndarray, Optional[np.ndarray], Dict]:
    """
    Encode categorical features using Label Encoding.

    Args:
        X (pd.DataFrame): Feature dataframe with categorical columns
        y (pd.DataFrame, optional): Target dataframe

    Returns:
        Tuple[np.ndarray, Optional[np.ndarray], Dict]:
            - Encoded features
            - Encoded target (if provided)
            - Dictionary of encoders for each column
    """
    logger.info("Encoding categorical features")

    # Initialize encoders dictionary
    encoders = {}

    # Create copies to avoid modifying originals
    X_encoded = X.copy()

    # Encode each feature
    for column in X.columns:
        # Initialize encoder
        encoder = LabelEncoder()

        # Fit and transform
        X_encoded[column] = encoder.fit_transform(X[column])

        # Store encoder
        encoders[column] = encoder

        # Log encoding information
        print(f"Encoded '{column}' with {len(encoder.classes_)} unique values")

    # Convert to numpy array
    X_array = X_encoded.values

    # Handle target encoding if provided
    y_array = None
    if y is not None:
        y_encoded = y.copy()
        target_col = y.columns[0]

        # Initialize encoder
        target_encoder = LabelEncoder()

        # Fit and transform
        y_encoded[target_col] = target_encoder.fit_transform(y[target_col])

        # Store encoder
        encoders[target_col] = target_encoder

        # Log target encoding mapping
        target_mapping = {label: idx for idx, label in enumerate(target_encoder.classes_)}
        print(f"Target encoding mapping: {target_mapping}")

        # Convert to numpy array
        y_array = y_encoded[target_col].values

    # Display sample of encoded data
    print("\nSample of encoded features:")
    print(X_encoded.head())

    return X_array, y_array, encoders
```

## Split Data

```python
def split_dataset(X: np.ndarray, y: np.ndarray, test_size: float = 0.2, random_state: int = 42) -> Tuple:
    """
    Split the dataset into training and testing sets.

    Args:
        X (np.ndarray): Feature array
        y (np.ndarray): Target array
        test_size (float): Proportion of data to use for testing
        random_state (int): Random seed for reproducibility

    Returns:
        Tuple: X_train, X_test, y_train, y_test
    """
    logger.info(f"Splitting data with test_size={test_size}")

    try:
        # Split the data with stratification to maintain class balance
        X_train, X_test, y_train, y_test = train_test_split(
            X,
            y,
            test_size=test_size,
            random_state=random_state,
            stratify=y  # Ensure balanced class distribution
        )

        # Log split information
        print(f"Training set: {X_train.shape[0]} samples")
        print(f"Testing set: {X_test.shape[0]} samples")

        # Verify class distribution in splits
        train_classes, train_counts = np.unique(y_train, return_counts=True)
        test_classes, test_counts = np.unique(y_test, return_counts=True)

        # Calculate percentages
        train_percentages = 100 * train_counts / len(y_train)
        test_percentages = 100 * test_counts / len(y_test)

        print("\nClass distribution in training set:")
        for cls, count, percentage in zip(train_classes, train_counts, train_percentages):
            print(f"Class {cls}: {count} samples ({percentage:.2f}%)")

        print("\nClass distribution in testing set:")
        for cls, count, percentage in zip(test_classes, test_counts, test_percentages):
            print(f"Class {cls}: {count} samples ({percentage:.2f}%)")

        return X_train, X_test, y_train, y_test

    except Exception as e:
        logger.error(f"Error splitting data: {str(e)}")
        raise
```

```python
def train_knn_model(X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray,
                   n_neighbors: int = 5) -> KNeighborsClassifier:
    """
    Train a K-Nearest Neighbors classifier.

    Args:
        X_train (np.ndarray): Training features
        y_train (np.ndarray): Training target
        X_test (np.ndarray): Testing features
        y_test (np.ndarray): Testing target
        n_neighbors (int): Number of neighbors to use

    Returns:
        KNeighborsClassifier: Trained KNN model
    """
    logger.info(f"Training KNN model with n_neighbors={n_neighbors}")

    # Initialize KNN model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)

    # Train the model
    knn.fit(X_train, y_train)

    # Make predictions
    y_pred = knn.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"KNN Model with n_neighbors={n_neighbors} metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    return knn
```

```python
def train_random_forest_model(X_train: np.ndarray, y_train: np.ndarray, X_test: np.ndarray, y_test: np.ndarray,
                             n_estimators: int = 100, random_state: int = 42) -> RandomForestClassifier:
    """
    Train a Random Forest classifier.

    Args:
        X_train (np.ndarray): Training features
        y_train (np.ndarray): Training target
        X_test (np.ndarray): Testing features
        y_test (np.ndarray): Testing target
        n_estimators (int): Number of trees in the forest
        random_state (int): Random seed for reproducibility

    Returns:
        RandomForestClassifier: Trained Random Forest model
    """
    logger.info(f"Training Random Forest model with n_estimators={n_estimators}")

    # Initialize Random Forest model
    rf = RandomForestClassifier(
        n_estimators=n_estimators,
        random_state=random_state
    )

    # Train the model
    rf.fit(X_train, y_train)

    # Make predictions
    y_pred = rf.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"Random Forest Model with n_estimators={n_estimators} metrics:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # Analyze feature importance if available
    if hasattr(rf, 'feature_importances_'):
        # Create dataframe for feature importance with original column names
        feature_names = list(X_processed.columns)
        importance_df = pd.DataFrame({
            'Feature': feature_names,
            'Importance': rf.feature_importances_
        }).sort_values('Importance', ascending=False)

        print("\nTop 10 important features:")
        for i, row in importance_df.head(10).iterrows():
            print(f"{row['Feature']}: {row['Importance']:.4f} ({row['Importance']*100:.2f}%)")

        # Visualize feature importance with better formatting
        plt.figure(figsize=(12, 8))
        top_features = importance_df.head(10)

        # Create horizontal bar chart with feature names
        ax = sns.barplot(x='Importance', y='Feature', data=top_features)

        # Add percentage labels
        for i, v in enumerate(top_features['Importance']):
            ax.text(v + 0.005, i, f'{v:.4f} ({v*100:.1f}%)', va='center')

        plt.title('Random Forest Feature Importance', fontsize=14)
        plt.xlabel('Importance Score', fontsize=12)
        plt.ylabel('Feature', fontsize=12)
        plt.tight_layout()
        plt.show()

    return rf
```

# Evaluations

```python
def evaluate_model(model: Any, model_name: str, X_test: np.ndarray, y_test: np.ndarray) -> Dict[str, float]:
    """
    Evaluate a classification model using various metrics.

    Args:
        model: Trained model to evaluate
        model_name (str): Name of the model
        X_test (np.ndarray): Test features
        y_test (np.ndarray): Test target

    Returns:
        Dict[str, float]: Dictionary of evaluation metrics
    """
    logger.info(f"Evaluating {model_name} model")

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate metrics
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred)
    }

    # Display metrics
    print(f"\n{model_name.upper()} EVALUATION METRICS:")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value:.4f}")

    # Display classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=['Edible', 'Poisonous']))

    # Display confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Edible', 'Poisonous'],
                yticklabels=['Edible', 'Poisonous'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()

    return metrics

def compare_models(models: Dict[str, Any], X_test: np.ndarray, y_test: np.ndarray) -> pd.DataFrame:
    """
    Compare performance of multiple models.

    Args:
        models (Dict[str, Any]): Dictionary of models to compare
        X_test (np.ndarray): Test features
        y_test (np.ndarray): Test target

    Returns:
        pd.DataFrame: DataFrame with model comparison
    """
    logger.info(f"Comparing {len(models)} models")

    # Evaluate each model
    results = {}
    for name, model in models.items():
        results[name] = evaluate_model(model, name, X_test, y_test)

    # Create comparison dataframe
    comparison = pd.DataFrame(results).T

    # Display comparison
    print("\nModel Comparison:")
    print(comparison)

    # Visualize comparison
    plt.figure(figsize=(10, 6))
    ax = comparison.plot(kind='bar', figsize=(10, 6))

    # Customize appearance
    plt.title('Model Performance Comparison', fontsize=14)
    plt.xlabel('Model', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.xticks(rotation=0)
    plt.ylim(0.8, 1.01)  # Set y-axis to start from 0.8 for better visibility of differences
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add value labels on top of bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%.4f', padding=3)

    plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.15), ncol=4)
    plt.tight_layout()
    plt.show()

    return comparison

# Evaluate and compare models
models_to_evaluate = {
    'KNN': knn_model,
    'Random Forest': rf_model
}
```
